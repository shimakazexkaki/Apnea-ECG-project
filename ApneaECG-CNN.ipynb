```python
#!wget -r -N -c -np https://physionet.org/files/apnea-ecg/1.0.0/
#!pip install wfdb
import wfdb
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.model_selection import StratifiedGroupKFold
from imblearn.over_sampling import SMOTE
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Conv1D,
    MaxPooling1D,
    Dense,
    Dropout,
    BatchNormalization,
    SpatialDropout1D,
    GlobalAveragePooling1D,
    Input,
    Bidirectional,
    LSTM,
    Multiply,
    Attention,
)
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras import regularizers

# ------------------------#
#     Data Loading        #
# ------------------------#
def load_signal_and_annotations(record_name):
    record = wfdb.rdsamp(record_name)
    annotation = wfdb.rdann(record_name, 'apn')

    signal = record[0][:, 0]  # Single lead signal (channel 0)
    fs = record[1]['fs']      # Sampling frequency
    labels = annotation.symbol
    times = annotation.sample / fs  # Convert sample indices to seconds

    return signal, labels, times, fs

def segment_signal_and_labels(signal, labels, times, fs, segment_duration=60):
    segment_length = int(segment_duration * fs)
    segments = []
    segment_labels = []

    for i, start_time in enumerate(times):
        start_idx = int(start_time * fs)
        end_idx = start_idx + segment_length

        if end_idx <= len(signal):
            segments.append(signal[start_idx:end_idx])
            segment_labels.append(1 if labels[i] == "A" else 0)

    return np.array(segments), np.array(segment_labels)

def augment_ecg_data(data, labels):
    augmented_data = []
    augmented_labels = []

    for i, signal in enumerate(data):
        augmented_data.append(signal)
        augmented_labels.append(labels[i])

        noise = np.random.normal(0, 0.01, signal.shape)
        augmented_data.append(signal + noise)
        augmented_labels.append(labels[i])

        scale_factor = np.random.uniform(0.9, 1.1)
        augmented_data.append(signal * scale_factor)
        augmented_labels.append(labels[i])

        augmented_data.append(np.flip(signal))
        augmented_labels.append(labels[i])

    return np.array(augmented_data), np.array(augmented_labels)

# ------------------------#
#   Build Model           #
# ------------------------#
def build_rnn_model(input_shape):
    l2_reg = regularizers.l2(1e-3)
    inputs = Input(shape=input_shape)

    x = Conv1D(16, 5, activation="relu", kernel_regularizer=l2_reg)(inputs)
    x = BatchNormalization()(x)
    x = MaxPooling1D(2)(x)
    x = SpatialDropout1D(0.2)(x)

    x = Conv1D(32, 5, activation="relu", kernel_regularizer=l2_reg)(x)
    x = BatchNormalization()(x)
    x = MaxPooling1D(2)(x)
    x = SpatialDropout1D(0.3)(x)

    x = Conv1D(64, 3, activation="relu", kernel_regularizer=l2_reg)(x)
    x = BatchNormalization()(x)
    x = MaxPooling1D(2)(x)
    x = SpatialDropout1D(0.4)(x)

    # Attention Layer
    attention_weights = Dense(64, activation="softmax")(x)
    x = Multiply()([x, attention_weights])

    # Bidirectional LSTM
    x = Bidirectional(LSTM(32, return_sequences=False))(x)
    x = Dropout(0.5)(x)

    outputs = Dense(1, activation="sigmoid")(x)

    model = Model(inputs, outputs)
    model.compile(
        loss="binary_crossentropy",
        optimizer=RMSprop(learning_rate=0.001),
        metrics=["accuracy", "AUC", "Precision", "Recall"],
    )
    return model

# ------------------------#
#     Cross-Validation    #
# ------------------------#
subjects = [f"a{i:02d}" for i in range(1, 21)] + [f"b{i:02d}" for i in range(1, 6)] + [f"c{i:02d}" for i in range(1, 11)]
print("Number of subjects (a/b/c):", len(subjects))

X_list, y_list, groups = [], [], []
for idx, subj in enumerate(subjects):
    record_name = f"/kaggle/working/physionet.org/files/apnea-ecg/1.0.0/{subj}"
    signal, labels, times, fs = load_signal_and_annotations(record_name)
    ecg_segments, segment_labels = segment_signal_and_labels(signal, labels, times, fs)
    X_list.append(ecg_segments)
    y_list.append(segment_labels)
    groups.extend([idx] * len(segment_labels))

X_all = np.vstack(X_list)
y_all = np.concatenate(y_list)
print("Total segments:", X_all.shape[0])
print("Apnea(1) ratio:", np.mean(y_all))

scaler = StandardScaler()
X_all = scaler.fit_transform(X_all)
X_all = np.expand_dims(X_all, axis=-1)

group_kfold = StratifiedGroupKFold(n_splits=5)
history_list = []
auc_scores, acc_scores = [], []

for fold_idx, (train_index, test_index) in enumerate(group_kfold.split(X_all, y_all, groups)):
    X_train, X_test = X_all[train_index], X_all[test_index]
    y_train, y_test = y_all[train_index], y_all[test_index]

    smote = SMOTE(random_state=42)
    X_train_resampled = X_train.reshape(X_train.shape[0], -1)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train_resampled, y_train)
    X_train_resampled = X_train_resampled.reshape(X_train_resampled.shape[0], X_train.shape[1], -1)

    X_train_augmented, y_train_augmented = augment_ecg_data(X_train_resampled, y_train_resampled)
    X_train_augmented = X_train_augmented.reshape(X_train_augmented.shape[0], X_train_resampled.shape[1], -1)

    model = build_rnn_model((X_all.shape[1], 1))
    model.summary()

    early_stopping = EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True, mode="min")
    reduce_lr = ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=2, min_lr=1e-6, verbose=1)

    val_size = int(0.1 * len(X_train_augmented))
    X_val, y_val = X_train_augmented[:val_size], y_train_augmented[:val_size]
    X_train_final, y_train_final = X_train_augmented[val_size:], y_train_augmented[val_size:]

    history = model.fit(
        X_train_final,
        y_train_final,
        validation_data=(X_val, y_val),
        epochs=50,
        batch_size=16,
        callbacks=[early_stopping, reduce_lr],
        verbose=1,
    )
    history_list.append(history)

    evaluation = model.evaluate(X_test, y_test, verbose=0)
    acc_scores.append(evaluation[1])

    y_pred = model.predict(X_test, verbose=0)
    auc_scores.append(roc_auc_score(y_test, y_pred))
    print(f"Fold {fold_idx+1}: Test Accuracy = {evaluation[1]:.4f}, AUC = {roc_auc_score(y_test, y_pred):.4f}")

# ------------------------#
#     Visualization       #
# ------------------------#
def visualize_performance(histories):
    metrics = ['loss', 'accuracy', 'AUC']
    for metric in metrics:
        plt.figure(figsize=(10, 6))
        for i, history in enumerate(histories):
            plt.plot(history.history[metric], label=f"Fold {i+1} Training {metric}")
            plt.plot(history.history[f"val_{metric}"], linestyle="--", label=f"Fold {i+1} Validation {metric}")
        plt.title(f"{metric.capitalize()} Over Epochs")
        plt.xlabel("Epochs")
        plt.ylabel(metric.capitalize())
        plt.legend()
        plt.grid(True)
        plt.show()

visualize_performance(history_list)
print(f"Average Accuracy: {np.mean(acc_scores):.4f}")
print(f"Average AUC: {np.mean(auc_scores):.4f}")
```
